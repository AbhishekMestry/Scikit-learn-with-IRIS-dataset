## k nearest neighbor algorithm for classification
Similar to human being.
Suppose we want to predict the specis of an unknown iris whose feature values are known.

### Steps
#### Step 1:
Choose a value for k (say k=5)
#### Step 2:
Select k neighbors in the training data, which are nearest to the unknown data
#### Step 3:
Out of selected neighbors, choose the more popular one as the response of the unknown iris
Example for k-1, 5



## Steps in building a k-nearest neighbor classification model in scikit learn
### Step 1: Import required class for building the model(estimator)

The model 'k nearest neighbor classification'is implemented in the 'kNearestClassifier' class in the 'neighbors' module. We need to import that first

```python
from sklearn.neighbors import KNeighborsClassifier
```

### Step 2: Instantiate the estimator

```python
knn=KNeighborsClassifier(n_neighbors =5)
```

### Step 3: Fit the model

```python
knn.fit(X_train,y_train)
```

### Step 4: Predict labels of test data using the model.

Apply knn.pred() function on test data

```python
y_pred=knn.predict(X_test)
print("Prediction of test data:",y_pred)
```

```python
y_test.head(10)
```

### Step 5:  Find accuracy of the model designed
Accurcy method helps

```python
## accuracy
from sklearn import metrics

print(metrics.accuracy_score(y_test,y_pred))
```
